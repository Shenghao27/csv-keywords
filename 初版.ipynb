```python
# 找到字元及字元位置
regex = re.compile(r'CCS通訊異常')
print(list(regex.finditer(text)))
print(type(list(regex.finditer(text))))
```

```python
# 轉成 list
regex = re.compile(r'CCS通訊異常')
CCS_list = list(regex.finditer(text))
    
print(CCS_list)
print(type(CCS_list))
```

```python
print(len(CCS_list))
print(CCS_list[0])
print(CCS_list[0].span())
```

```python
# 取出 Span list
Span_list = []

for i in CCS_list:
    Span=i.span()
    Span_list.append(Span)
print(Span_list)
print(type(Span_list))
# print(type(str(Span_list)))
```

```python
# 轉成 int 做元素加減，才能取得異常訊息位置
key_3 = []
key_4 = []

for i in Span_list:
    key_1 = str(i)
    key_2 = key_1.strip('()').split(',')
    key_3.append(int(key_2[0]))  #需要轉成 'int'
    key_4.append(int(key_2[1]))  #需要轉成 'int'


print(key_3)
print(type(key_3))
print(key_4)
```

```python
print(CCS_numbers)
print(type(CCS_numbers))
print(len(CCS_numbers))
```

```python
a = CCS_numbers  # 新增變數 a，內容是使用者輸入的多個數字，數字以逗號分隔

output = 0            # 設定 output 從 0 開始
for i in a:           # 使用 for 迴圈，依序取出 b 串列的每個項目
    output += int(i)  # 將 output 的數值加上每個項目 ( 使用 int 將項目轉換成數字 )

print(f'數字總和為：{output}')
```
